```python
import xport
import numpy as np
import os
import pandas as pd
```

```python
# Function to convert XPT to CSV file
def XPT_TO_CSV(filename):
  with open(f'./Dataset/xpt files/{filename}', 'rb') as f:
    df = xport.to_dataframe(f)
  print (df.columns.values)
  df.to_csv(f'./Dataset/csv files/{filename[:-4]}.csv')
```

```python
xpt_files = os.listdir('./Dataset/xpt files')
xpt_files.sort()
xpt_files
```

```python
for i in xpt_files:
  XPT_TO_CSV(i)
```

```python
def extract_seqn(list_2, list_):
    bool_list = []
    for i in list_2:
        if i in list_:
            bool_list.append(True)
        else:
            bool_list.append(False)
    return bool_list
```
