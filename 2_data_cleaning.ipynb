```python
df_bmx = df[['BMDBMIC', 'BMXBMI']]
```

```python
df_bmx[df_bmx['BMDBMIC'].isnull() & ~(df_bmx['BMXBMI'].isnull())]
```

```python
df_bmx[df_bmx['BMXBMI'].isnull()]
```

```python
def BMI_CAT(BMI):
    if(BMI==0):
        return np.nan
    elif(BMI<18.5):
        return 1
    elif(BMI>=18.5 and BMI<25):
        return 2
    elif(BMI>=25 and BMI<30):
        return 3
    else:
        return 4
```

```python
df = df[~(df['RIAGENDR'].isnull())]
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
BMC = []
BMI = df_bmx['BMXBMI'].fillna(0)
```

```python
for i in BMI:
    BMC.append(BMI_CAT(i))
```

```python
BMC = np.asarray(BMC)
df['BMDBMIC'] = BMC 
```
